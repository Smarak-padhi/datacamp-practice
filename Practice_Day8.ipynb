# Finding Missing Values

## Overview

Missing values are everywhere, and you don't want them interfering with your work. Some functions ignore missing data by default, but that's not always the behavior you might want. Some functions can't handle missing values at all, so these values need to be taken care of before you can use them. If you don't know where your missing values are, or if they exist, you could make mistakes in your analysis.

In this exercise, you'll determine if there are missing values in the dataset, and if so, how many.

**Libraries:**
- `pandas` imported as `pd`

**Dataset:** `avocados_2016`

## Instructions

1. Print a DataFrame that shows whether each value in `avocados_2016` is missing or not.
2. Print a summary that shows whether any value in each column is missing or not.
3. Create a bar plot of the total number of missing values in each column.

```python
# Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt

# Check individual values for missing values
print(avocados_2016.isna())

# Check each column for missing values
print(avocados_2016.isna().any())

# Bar plot of missing values by variable
print(avocados_2016.isna().sum().plot(kind="bar"))

# Show plot
plt.show()
```

# CSV to DataFrame

## Overview

You work for an airline, and your manager has asked you to perform a competitive analysis of how often passengers flying on other airlines are involuntarily bumped from their flights.

The file `airline_bumping.csv` contains data about passengers who were involuntarily denied boarding in 2016 and 2017.

**Library:**
- `pandas` imported as `pd`

**File:**
- `airline_bumping.csv`

## Instructions

1. Read `airline_bumping.csv` into a DataFrame called `airline_bumping`.
2. Print the first few rows of `airline_bumping`.
3. Group the data by `airline`.
4. Select the `nb_bumped` and `total_passengers` columns.
5. Calculate the sum for both years and store the result as `airline_totals`.
6. Create a column called `bumps_per_10k` containing the number of passengers bumped per 10,000 passengers.
7. Print `airline_totals`.

```python
# Read CSV as a DataFrame
airline_bumping = pd.read_csv("airline_bumping.csv")

# Print the first few rows
print(airline_bumping.head())

# Calculate totals for each airline
airline_totals = airline_bumping.groupby("airline")[
    ["nb_bumped", "total_passengers"]
].sum()

# Calculate passengers bumped per 10,000 passengers
airline_totals["bumps_per_10k"] = (
    airline_totals["nb_bumped"]
    / airline_totals["total_passengers"]
    * 10000
)

# Print the results
print(airline_totals)
```

You're almost there! To make things easier to read, you'll need to sort the data and export it to a CSV so that your colleagues can read it.

**Library:**
