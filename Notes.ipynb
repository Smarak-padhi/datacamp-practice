# 📆 3. Partial Date Slicing

Pandas understands partial date strings.

## Entire Year

```python
dogs.loc["2015"]
```

Returns every row in **2015**.

---

## Range of Years

```python
dogs.loc["2014":"2016"]
```

Equivalent to

```python
dogs.loc["2014-01-01":"2016-12-31"]
```

Returns every row from 2014 through 2016.

---

## Entire Month

```python
dogs.loc["2015-04"]
```

Returns every row from April 2015.

---

## Month Range

```python
dogs.loc["2014-08":"2015-04"]
```

Returns

- August 2014
- September 2014
- ...
- April 2015

---

# 💡 Why Partial Dates Work

Once the index is a **DatetimeIndex**, pandas automatically interprets

```python
"2014"
```

as

```
2014-01-01
↓
2014-12-31
```

and

```python
"2015-04"
```

as

```
2015-04-01
↓
2015-04-30
```

# ⚖️ `.loc` vs `.iloc`

| `.loc` | `.iloc` |
|---------|----------|
| Uses labels | Uses integer positions |
| `dogs.loc["Bella"]` | `dogs.iloc[0]` |
| End included | End excluded |
| Great for dates | Uses positions only |
