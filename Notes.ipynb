# Page 31 – Working with Pivot Tables

## What is a Pivot Table?

A **pivot table** is used to summarize and organize data.

Instead of looking through hundreds of rows manually, a pivot table groups the data and calculates statistics such as:

- Mean (Average)
- Sum
- Count
- Minimum
- Maximum

### Why do we use Pivot Tables?

Imagine a dataset containing information about 100 dogs.

Without a pivot table:
- You would need to filter each breed manually.
- Then calculate averages one by one.

With a pivot table:
- Pandas groups the data automatically.
- Calculates the required statistic in one line of code.

**Think of a pivot table as:**

> "Group my data and summarize it."
