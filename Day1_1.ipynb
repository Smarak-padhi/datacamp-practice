# What is a One-to-Many Relationship?

A **One-to-Many** relationship occurs when:

> One row in the **left table** matches **multiple rows** in the **right table**.

Unlike a One-to-One relationship, the merge key appears **only once** in one table but **multiple times** in the other.

---

## Definition

**One-to-Many = Every row in the left table is related to one or more rows in the right table.**

Example:

```
Left Table (Customers)

Customer ID
-----------
101
102
103
```

```
Right Table (Orders)

Customer ID
-----------
101
101
101
102
103
103
```

Customer **101** has placed three orders.

---

# Visual Representation

```
Customer 101
     │
     ├── Order 1
     ├── Order 2
     └── Order 3

Customer 102
     │
     └── Order 4

Customer 103
     │
     ├── Order 5
     └── Order 6
```

One customer can have many orders.

---

# One-to-One vs One-to-Many

## One-to-One

Each key appears **once** in both tables.

Example:

### Customers

| Customer ID |
|-------------|
| 101 |
| 102 |
| 103 |

### Passports

| Customer ID |
|-------------|
| 101 |
| 102 |
| 103 |

Every customer has exactly one passport.

---

## One-to-Many

A key appears once in one table and multiple times in another.

### Customers

| Customer ID |
|-------------|
| 101 |
| 102 |
| 103 |

### Orders

| Customer ID |
|-------------|
| 101 |
| 101 |
| 101 |
| 102 |
| 103 |
| 103 |

Customer 101 owns multiple orders.

---

# Real-Life Examples

| One | Many |
|-----|------|
| Customer | Orders |
| Student | Exam Scores |
| Department | Employees |
| Country | Cities |
| Product | Sales |
| Ward | Business Licenses |

Whenever one record owns multiple records, it is a One-to-Many relationship.

- `.merge()`
- `on=`
- optional `suffixes=`

Only the relationship between the tables changes.

# Number of Rows After Merge

This is the biggest difference between One-to-One and One-to-Many.

## One-to-One

```
50 rows

+

50 rows

↓

50 rows
```

The number of rows usually stays the same.

---

## One-to-Many

```
50 wards

+

10,000 businesses

↓

10,000 merged rows
```

Each business receives information from its ward.

The merged table becomes much larger.

---

# Why Do Rows Increase?

Suppose:

```
Customers

101
102
```

Orders

```
101
101
101
102
```

After merging:

```
101 Order A
101 Order B
101 Order C
102 Order D
```

Customer **101** is repeated because it has three matching rows.

# How to Identify a One-to-Many Relationship

Ask yourself:

**Does one value in the left table match multiple rows in the right table?**

If **YES**, it is One-to-Many.

Examples:

✅ Customer → Orders

✅ Student → Marks

✅ Department → Employees

✅ Ward → Business Licenses

---

# One-to-One vs One-to-Many Summary

| Feature | One-to-One | One-to-Many |
|----------|------------|-------------|
| Left key unique | ✅ Yes | ✅ Yes |
| Right key unique | ✅ Yes | ❌ No |
| Duplicate rows after merge | ❌ No | ✅ Yes |
| Rows after merge | Usually same | Usually increases |
| Common example | Customer → Passport | Customer → Orders |

---

# Memory Trick

Think of a parent and children.

```
Parent

    │
    ├── Child 1
    ├── Child 2
    ├── Child 3
    └── Child 4
```

One parent can have many children.

Similarly,

```
Customer

    │
    ├── Order 1
    ├── Order 2
    ├── Order 3
    └── Order 4
```

One customer can have many orders.

---

# Key Takeaways

- One-to-Many means one record matches multiple records.
- The merge syntax stays exactly the same.
- The merge key is unique in one table but repeated in the other.
- Rows usually increase after merging.
- Duplicate rows in the merged table are expected.
- Common examples include Customers → Orders, Students → Marks, and Wards → Business Licenses.

# Why Merge Multiple DataFrames?

Sometimes all the information you need is spread across more than two DataFrames.

Instead of merging only two tables, you can merge several DataFrames together to build one complete dataset.

Example:

```
Customers
      +
Orders
      +
Payments

↓

Complete Customer Data
```

---

# Merging Two DataFrames

Basic syntax:

```python
merged_df = df1.merge(df2, on="column_name")
```

Example:

```python
wards_census = wards.merge(census, on="ward")
```

---

# Merging More Than Two DataFrames

You can merge another DataFrame onto an already merged DataFrame.

Example:

```python
df = df1.merge(df2, on="id")
df = df.merge(df3, on="id")
```

You simply continue merging until all required DataFrames are combined.

---

# Example

Suppose you have three DataFrames.

### Students

| id | name |
|----|------|
|1|Alice|
|2|Bob|

### Scores

| id | marks |
|----|-------|
|1|90|
|2|85|

### Attendance

| id | attendance |
|----|------------|
|1|95%|
|2|88%|

Merge them one by one:

```python
student_data = students.merge(scores, on="id")

student_data = student_data.merge(attendance, on="id")
```

Final result:

| id | name | marks | attendance |
|----|------|-------|------------|
|1|Alice|90|95%|
|2|Bob|85|88%|

---

# Merging on Multiple Columns

Sometimes one column is not enough to uniquely identify a row.

Instead of merging on a single column:

```python
on="zip"
```

merge using multiple columns:

```python
on=["address", "zip"]
```

Notice that multiple columns are passed as a **list**.

---

# Why Merge on Multiple Columns?

Suppose two businesses share the same ZIP code.

### Grants

| business | address | zip |
|----------|---------|-----|
|Reggie's|123 Main St|60616|

### Licenses

| business | address | zip |
|----------|---------|-----|
|Reggie's|123 Main St|60616|
|Pizza Hut|500 Oak St|60616|
|Subway|800 Pine St|60616|

If you merge only on:

```python
on="zip"
```

Every business with ZIP **60616** will match, creating incorrect duplicate rows.

---

# Correct Approach

Merge using both:

```python
grants.merge(
    licenses,
    on=["address", "zip"]
)
```

Now pandas matches rows only when:

- Address matches
- ZIP code matches

This produces a much more accurate merge.

---

# Choosing the Correct Merge Key

Always choose columns that uniquely identify a record.

Good merge keys:

- Customer ID
- Employee ID
- Vehicle ID
- Student ID

Sometimes one column is enough.

Sometimes multiple columns are needed.

---

# Merging Process

```
DataFrame 1
      │
      ▼
Merge
      │
      ▼
DataFrame 2
      │
      ▼
Merged DataFrame
      │
      ▼
Merge Again
      │
      ▼
DataFrame 3
      │
      ▼
Final DataFrame
```

---

# Common Mistakes

### Merging on the wrong column

```python
on="zip"
```

This may create incorrect matches because many records can share the same ZIP code.

---

### Forgetting that multiple columns can be used

Instead of:

```python
on="zip"
```

use:

```python
on=["address", "zip"]
```

when both columns are needed to uniquely identify a record.

---

# Key Takeaways

- You can merge more than two DataFrames.
- Merge DataFrames one at a time.
- Use `.merge()` repeatedly.
- Merge on multiple columns by passing a list.
- Always choose merge keys that uniquely identify records.
- Using the wrong merge key can produce duplicate or incorrect matches.

# Exercise 3 — Total Riders in a Month

## Problem

Three DataFrames are provided:

- `ridership` – Number of rides for each station each day.
- `cal` – Calendar information (`year`, `month`, `day`, `day_type`).
- `stations` – Station details (`station_id`, `station_name`, `location`).

### Tasks

1. Merge `ridership` and `cal` using:
   - `year`
   - `month`
   - `day`

2. Merge the result with the `stations` DataFrame using:
   - `station_id`

3. Create a filter to select:
   - Month = **July (`7`)**
   - Day Type = **Weekday**
   - Station Name = **Wilson**

4. Print the **total number of rides**.

# Solution

```python
# Merge ridership and calendar
ridership_cal = ridership.merge(
    cal,
    on=["year", "month", "day"]
)

# Merge with stations
ridership_cal_stations = ridership_cal.merge(
    stations,
    on="station_id"
)

# Filter the required rows
filter_criteria = (
    (ridership_cal_stations["month"] == 7) &
    (ridership_cal_stations["day_type"] == "Weekday") &
    (ridership_cal_stations["station_name"] == "Wilson")
)

# Print total rides
print(
    ridership_cal_stations.loc[
        filter_criteria,
        "rides"
    ].sum()
)
```

### Explanation

- Merge `ridership` and `cal` using the date columns.
- Merge the result with `stations` using `station_id`.
- Filter only **Weekdays in July** for the **Wilson** station.
- Select the `rides` column and calculate the total using `.sum()`.

# Exercise 4 — Three Table Merge

## Problem

Three DataFrames are provided:

- `licenses`
- `zip_demo`
- `wards`

### Tasks

1. Merge `licenses` with `zip_demo` using the `zip` column.
2. Merge the resulting DataFrame with `wards` using the `ward` column.
3. Store the result in `licenses_zip_ward`.
4. Group the data by `alderman`.
5. Calculate the median `income` for each alderman.

# Solution

```python
# Merge licenses with zip_demo
licenses_zip_ward = licenses.merge(
    zip_demo,
    on="zip"
).merge(
    wards,
    on="ward"
)

# Calculate median income for each alderman
print(
    licenses_zip_ward.groupby("alderman")
                     .agg({"income": "median"})
)
```

### Explanation

- Merge `licenses` and `zip_demo` using `zip`.
- Merge the result with `wards` using `ward`.
- Group the data by `alderman`.
- Calculate the median value of the `income` column using `.agg()`.

# Exercise 5 — One-to-Many Merge with Multiple Tables

## Problem

Three DataFrames are provided:

- `land_use`
- `census`
- `licenses`

### Tasks

1. Merge `land_use` and `census` using the `ward` column.
2. Merge the result with `licenses` using the `ward` column.
3. Use the suffixes:
   - `_cen` for the left table
   - `_lic` for the right table
4. Group the merged DataFrame by:
   - `ward`
   - `pop_2010`
   - `vacant`
5. Count the number of `account`s in each group.
6. Sort the results by:
   - `vacant` (Descending)
   - `account` (Ascending)
   - `pop_2010` (Ascending)
7. Print the first five rows.

# Solution

```python
# Merge the three DataFrames
land_cen_lic = (
    land_use
    .merge(census, on="ward")
    .merge(
        licenses,
        on="ward",
        suffixes=("_cen", "_lic")
    )
)

# Group and count accounts
pop_vac_lic = (
    land_cen_lic
    .groupby(
        ["ward", "pop_2010", "vacant"],
        as_index=False
    )
    .agg({"account": "count"})
)

# Sort the results
sorted_pop_vac_lic = pop_vac_lic.sort_values(
    ["vacant", "account", "pop_2010"],
    ascending=[False, True, True]
)

# Display the first five rows
print(sorted_pop_vac_lic.head())
```

### Explanation

- Merge `land_use` with `census` on `ward`.
- Merge the result with `licenses` using `ward`.
- Add suffixes to distinguish duplicate columns.
- Group by `ward`, `pop_2010`, and `vacant`.
- Count the number of business `account`s.
- Sort the results and display the first five rows.
