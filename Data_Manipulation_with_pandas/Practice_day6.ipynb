# Pivot Temperature by Country, City, and Year

## Objective

We want to reorganize the temperature dataset so that:

- Each row represents a `country` and `city`
- Each column represents a `year`
- Each value shows the average temperature, `avg_temp_c`

---

## Extracting the Year from a Date Column

The `date` column contains complete dates.

We can extract only the year using:

```python
temperatures["year"] = temperatures["date"].dt.year

````python
# Add a year column to temperatures
temperatures["year"] = temperatures["date"].dt.year

# Pivot avg_temp_c by country and city vs year
temp_by_country_city_vs_year = temperatures.pivot_table(values="avg_temp_c", index=["country", "city"], columns="year")

# See the result
print(temp_by_country_city_vs_year)
````

# Subsetting Pivot Tables

## Objective

Use `.loc[]` on `temp_by_country_city_vs_year` to create the following subsets:

1. From Egypt to India.
2. From Egypt, Cairo to India, Delhi.
3. From Egypt, Cairo to India, Delhi, and years 2005 to 2010.

---

## Solution

```python
# From Egypt to India
temp_by_country_city_vs_year.loc["Egypt":"India"]

# From Egypt, Cairo to India, Delhi
temp_by_country_city_vs_year.loc[
    ("Egypt", "Cairo"):("India", "Delhi")
]

# From Egypt, Cairo to India, Delhi, and years 2005 to 2010
temp_by_country_city_vs_year.loc[
    ("Egypt", "Cairo"):("India", "Delhi"),
    2005:2010
]
```---

# 2. Subset from Egypt, Cairo to India, Delhi

```python
temp_by_country_city_vs_year.loc[
    ("Egypt", "Cairo"):("India", "Delhi")
]
```

### Explanation

The DataFrame has a **MultiIndex**:

```
Country
    City
```

So each row is identified by a tuple:

```python
(country, city)
```

Here,

```python
("Egypt", "Cairo")
```

is the starting row, and

```python
("India", "Delhi")
```

is the ending row.

Everything between these two rows is selected.

---

# 3. Subset Rows and Columns Together

```python
temp_by_country_city_vs_year.loc[
    ("Egypt", "Cairo"):("India", "Delhi"),
    "2005":"2010"
]
```

### Explanation

`.loc[]` has the syntax:

```python
df.loc[rows, columns]
```

### Rows

```python
("Egypt", "Cairo"):("India", "Delhi")
```

Selects rows from Egypt → India.

### Columns

```python
"2005":"2010"
```

Selects only the years from **2005 to 2010**.

---

# General Syntax of `.loc`

```python
df.loc[row_selection, column_selection]
```

Examples:

```python
# Only rows
df.loc["A":"D"]

# Only columns
df.loc[:, "2005":"2010"]

# Rows and columns
df.loc["A":"D", "2005":"2010"]
```

The colon (`:`) means **from ... to ... (inclusive)**.

---

# Important Note

If the column names are integers (years stored as numbers), write:

```python
2005:2010
```

If they are strings, write:

```python
"2005":"2010"
```

DataCamp stores the year columns as **integers**, so the correct solution is:

```python
temp_by_country_city_vs_year.loc[
    ("Egypt", "Cairo"):("India", "Delhi"),
    2005:2010
]
```

temp_by_country_city_vs_year.loc["Egypt":"India"]

temp_by_country_city_vs_year.loc[

]

temp_by_country_city_vs_year.loc[

    2005:2010

```

---

# Key Takeaways

- `.loc[]` selects data **by labels**.
- A pivot table is still a **DataFrame**, so all DataFrame indexing methods work.
- Use tuples like `(country, city)` when working with a **MultiIndex**.
- `.loc[rows, columns]` lets you filter both rows and columns at the same time.
- Label slicing in `.loc` is **inclusive**, meaning both the start and end labels are included.

# Calculating Summary Statistics from a Pivot Table

## Question

1. Calculate the mean temperature for each year and assign it to `mean_temp_by_year`.
2. Filter `mean_temp_by_year` for the year with the highest mean temperature.
3. Calculate the mean temperature for each city across all years and assign it to `mean_temp_by_city`.
4. Filter `mean_temp_by_city` for the city with the lowest mean temperature.


```

````python
# Get the worldwide mean temp by year
mean_temp_by_year = temp_by_country_city_vs_year.mean()

# Filter for the year that had the highest mean temp
print(mean_temp_by_year[mean_temp_by_year == mean_temp_by_year.max()])

# Get the mean temp by city
mean_temp_by_city = temp_by_country_city_vs_year.mean(axis="columns")

# Filter for the city that had the lowest mean temp
print(mean_temp_by_city[mean_temp_by_city == mean_temp_by_city.min()])
````
