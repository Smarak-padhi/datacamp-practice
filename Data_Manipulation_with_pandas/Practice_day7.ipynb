# Changes in Sales Over Time

## Overview

Line plots are designed to visualize the relationship between two numeric variables, where each data value is connected to the next one. They are especially useful for visualizing changes in a number over time since each time point is naturally connected to the next.

In this exercise, you'll visualize the change in avocado sales over three years.

**Dataset:** `avocados`  
**Library:** `pandas` (imported as `pd`)

## Instructions

1. Get the total number of avocados sold on each date. The DataFrame has two rows for each date:
   - One for **organic**
   - One for **conventional**

   Save the result as `nb_sold_by_date`.

2. Create a **line plot** of the number of avocados sold.

3. Display the plot.

```python
# Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt

# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby("date")["nb_sold"].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(x = "nb_sold",
                    y = "date",
                    kind = "line")

# Show the plot
plt.show()
```

# Avocado Supply and Demand

## Overview

Scatter plots are ideal for visualizing relationships between numerical variables. In this exercise, you'll compare the number of avocados sold to the average price to see if they are related. If they are, you may be able to use one variable to predict the other.

**Libraries:**
- `matplotlib.pyplot` imported as `plt`
- `pandas` imported as `pd`

**Dataset:** `avocados`

## Instructions

1. Create a **scatter plot** with:
   - `nb_sold` on the **x-axis**
   - `avg_price` on the **y-axis**
2. Set the title of the plot to:
   - `"Number of avocados sold vs. average price"`

```python
# Scatter plot of avg_price vs. nb_sold with title
avocados.plot(x="nb_sold", y="avg_price", kind="scatter", title="Number of avocados sold vs. average price")

# Show the plot
plt.show()
```
