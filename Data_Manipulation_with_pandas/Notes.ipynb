### What are Missing Values?

A **missing value** is a piece of data that is not available or has not been recorded.

Example:

| Name | Breed | Weight (kg) |
|------|-------|-------------|
| Bella | Labrador | ? |
| Charlie | Poodle | 23 |
| Cooper | Schnauzer | ? |

The `?` indicates that the value is unknown.

---

## Why do Missing Values Occur?

Missing values can occur because:

- Data was not entered.
- Data was lost during collection.
- The measurement was not taken.
- Errors occurred while importing data.

---

## Why are Missing Values Important?

Missing values can:

- Produce incorrect calculations.
- Cause errors in machine learning models.
- Affect graphs and summary statistics.

Therefore, they should always be checked before analyzing data.

---

## Key Point

Missing values are **not the same as zero**.

```
Missing Value (NaN)
≠
0
```

A missing value means **unknown**, while 0 is an actual recorded value.

---

# Page 22 - Removing Missing Values

## Topic: Removing Rows with Missing Data

Sometimes, the easiest solution is to remove rows containing missing values.

Pandas provides the `.dropna()` function for this.

---

## Syntax

```python
dogs.dropna()
```

---

## What does `.dropna()` do?

It removes every row that contains **at least one missing value (NaN).**

---

## Example

Original DataFrame

| Name | Weight |
|------|--------|
| Bella | NaN |
| Charlie | 24 |
| Cooper | NaN |
| Max | 29 |

After

```python
dogs.dropna()
```

Output

| Name | Weight |
|------|--------|
| Charlie | 24 |
| Max | 29 |

Rows containing NaN are removed.

---

## When should you use `.dropna()`?

Use it when:

- Only a few rows contain missing values.
- Removing those rows will not affect your analysis.

---

## When should you NOT use it?

Avoid using `.dropna()` if:

- Many rows contain missing values.
- Removing them would result in losing too much data.

---

## Important Note

`.dropna()` returns a **new DataFrame**.

The original DataFrame remains unchanged unless you assign it back.

Example

```python
dogs = dogs.dropna()
```

or

```python
dogs.dropna(inplace=True)
```

---

# Page 23 - Replacing Missing Values

## Topic: Filling Missing Values

Instead of removing rows,

you can replace missing values with another value.

---

## Syntax

```python
dogs.fillna(0)
```

---

## What does `.fillna()` do?

It replaces every missing value (`NaN`) with the value you specify.

Here,

```
NaN

↓

0
```

---

## Example

Original DataFrame

| Name | Weight |
|------|--------|
| Bella | NaN |
| Charlie | 24 |
| Cooper | NaN |

After

```python
dogs.fillna(0)
```

Output

| Name | Weight |
|------|--------|
| Bella | 0 |
| Charlie | 24 |
| Cooper | 0 |

---

## Why use `.fillna()`?

Sometimes you don't want to lose data by deleting rows.

Instead,

you replace missing values with something meaningful.

Examples

Replace with

```python
0
```

Useful when missing means **none**.

---

Replace with

```python
dogs["weight_kg"].mean()
```

Useful for numerical columns.

---

Replace with

```python
"Unknown"
```

Useful for text columns.

---

## Important Note

Like `.dropna()`,

`.fillna()` returns a new DataFrame.

To save the changes

```python
dogs = dogs.fillna(0)
```

or

```python
dogs.fillna(0, inplace=True)
```

---

# Choosing Between dropna() and fillna()

## Use `.dropna()` when

- Very few rows have missing values.
- Removing them won't affect the dataset.

Example

```
1000 rows

↓

Only 3 rows contain NaN

↓

Safe to remove.
```

---

## Use `.fillna()` when

- Many rows contain missing values.
- You want to preserve the dataset.

Example

```
1000 rows

↓

300 rows contain NaN

↓

Replacing is better than deleting.
```

Instead of creating a DataFrame **row by row**, we can also create it **column by column**.

## What is a Dictionary of Lists?

A dictionary where:

- Each **value** is a list containing all the values for that column.

General Structure

```python

    "Column1": [value1, value2, value3],

    "Column3": [value1, value2, value3]

```

---

```python
dict_of_lists = {
    "name": ["Ginger", "Scout"],

    "height_cm": [22, 59],

    "date_of_birth": ["2019-03-14", "2019-05-09"]
}
```

---

## Understanding the Structure

```
"name"

↓

Column Name

↓

["Ginger", "Scout"]

↓

Entire Column
```

Every key stores one complete column.

---

## Memory Trick

```
One Key

↓

One Column

↓

One List
```

Unlike a list of dictionaries,

where one dictionary represented one row,

here one key represents one column.
