# Why Do We Merge DataFrames?

Real-world data is rarely stored in one large table.

Instead, information is split into multiple DataFrames.

For example:

### DataFrame 1 (Employees)

| employee_id | employee_name |
|--------------|---------------|
|101|Alice|
|102|Bob|

### DataFrame 2 (Salaries)

| employee_id | salary |
|--------------|---------|
|101|50000|
|102|65000|

To know each employee's salary, we need to combine these two DataFrames.

This process is called **merging** (or joining).

# What is a Key?

A key is the column that exists in both DataFrames and is used to match rows.

Example

Employees

| employee_id | name |
|--------------|------|
|1|Alice|
|2|Bob|

Salary

| employee_id | salary |
|--------------|---------|
|1|50000|
|2|70000|

The key is:

```
employee_id
```

---

# Inner Join

An **Inner Join** returns **only the rows that exist in BOTH DataFrames**.

Think of it as finding the common records between two tables.

Visualization

```
Left DataFrame       Right DataFrame

      ○────────○
     /          \
    /   Common   \
    \   Records  /
     \          /
      ○────────○

Only the overlapping area is returned.
```

---

# Syntax

```python
df1.merge(df2, on="column_name")
```

or

```python
pd.merge(df1, df2, on="column_name")
```

Both produce the same result.

---

# Example

```python
wards_census = wards.merge(
    census,
    on="ward"
)
```

Explanation

- `wards` → Left DataFrame
- `census` → Right DataFrame
- `on="ward"` → Match rows using the `ward` column

---

# Understanding the Output

Before merging

```
wards
(50, 4)

census
(50, 6)
```

After merging

```
(50, 9)
```

Why 9 columns?

```
4 + 6 = 10

One column ("ward") is shared

10 - 1 = 9
```

# Duplicate Columns

Sometimes both DataFrames contain columns with the same name.

Example

Both have

```
address
zip
```

Pandas cannot keep two columns with identical names.

So it automatically renames them.

```
address_x
address_y

zip_x
zip_y
```

---

# What do the suffixes mean?

```
_x
```

Column from the left DataFrame.

```
_y
```

Column from the right DataFrame.

Example

```
address_x
```

→ Address from the Ward DataFrame.

```
address_y
```

→ Address from the Census DataFrame.

---

# Custom Suffixes

Instead of `_x` and `_y`, we can provide our own names.

```python
wards.merge(
    census,
    on="ward",
    suffixes=("_ward", "_cen")
)
```

Result

Instead of

```
address_x
address_y
```

we get

```
address_ward
address_cen
```

which is much easier to understand.

---

# Why Use Custom Suffixes?

Custom suffixes make the merged DataFrame more readable.

Instead of guessing:

```
zip_x
```

you immediately know:

```
zip_ward
```

belongs to the Ward DataFrame.

---

# Summary

- Joining and merging mean the same thing.
- DataFrames are merged using a common column called the **key**.
- An Inner Join returns only matching rows from both DataFrames.
- The `on` parameter specifies the matching column.
- Duplicate column names automatically receive `_x` and `_y`.
- Use `suffixes=` to provide meaningful names for duplicate columns.

---

# Common Mistakes

### Forgetting the key

❌

```python
wards.merge(census)
```

Pandas tries to guess the common columns, which may not be what you want.

Always specify the key.

✅

```python
wards.merge(
    census,
    on="ward"
)
```

---

### Confusing Inner Join with Left Join

Remember:

Inner Join only keeps matching rows.

Rows that don't exist in both DataFrames are removed.

---

# Interview Tip

A very common interview question is:

> What is the difference between an Inner Join and a Left Join?

Answer:

- **Inner Join** keeps only matching rows.
- **Left Join** keeps all rows from the left DataFrame, even if there is no match in the right DataFrame.

You'll learn Left Joins in the next lesson.

# Exercise 2 — Your First Inner Join

## Objective

Determine the most popular type of fuel used by taxis in Chicago by merging two DataFrames.

---

## Problem Statement

You have been tasked with finding the most common **fuel type** used by Chicago taxis.

Two DataFrames are provided:

- `taxi_owners` — Contains information about taxi owners.
- `taxi_veh` — Contains information about each taxi vehicle.

To complete the analysis:

1. Merge the two DataFrames using the **`vid`** column.
2. Use custom suffixes to distinguish overlapping column names.
3. Count the occurrences of each `fuel_type` to identify the most common one.

---

## Instructions

### Step 1

Merge `taxi_owners` with `taxi_veh` using:

- Merge key: `vid`
- Store the result in `taxi_own_veh`

---

### Step 2

Use custom suffixes for overlapping column names:

- Left DataFrame → `_own`
- Right DataFrame → `_veh`

---

### Step 3

Select the `fuel_type` column from the merged DataFrame and display the frequency of each fuel type using the `value_counts()` method.

---

## Expected Outcome

The output should display the number of taxis for each fuel type, allowing you to determine which fuel type is the most common.

# Solutions

## Solution

✅ **Correct Answer**

```python
on="vid"
```

### Explanation

A merge should always be performed using a column that:

- Exists in **both** DataFrames.
- Represents the **same information**.
- Can uniquely relate rows between the two tables.

The `vid` (**Vehicle ID**) column satisfies all three conditions.

The other columns cannot be used because:

- `rid` only exists in `taxi_owners`.
- `year` only exists in `taxi_veh`.
- `zip` only exists in `taxi_owners`.

Therefore, the correct merge key is:

```python
taxi_owners.merge(taxi_veh, on="vid")
```

## Solution

```python
# Merge the taxi_owners and taxi_veh tables
taxi_own_veh = taxi_owners.merge(
    taxi_veh,
    on="vid",
    suffixes=("_own", "_veh")
)

# Count the number of taxis for each fuel type
print(taxi_own_veh["fuel_type"].value_counts())
```

---

## Explanation

### Merge the DataFrames

```python
taxi_owners.merge(
    taxi_veh,
    on="vid"
)
```

- Combines both DataFrames.
- Matches rows using the `vid` (Vehicle ID) column.
- Only rows with matching `vid` values are included because `merge()` performs an **Inner Join** by default.

---

### Apply Custom Suffixes

```python
suffixes=("_own", "_veh")
```

Both DataFrames contain columns such as:

- `owner`

To avoid duplicate column names, pandas renames them as:

```
owner_own
owner_veh
```

making it clear which DataFrame each column originated from.

---

### Count Fuel Types

```python
taxi_own_veh["fuel_type"].value_counts()
```

- Selects the `fuel_type` column.
- Counts how many times each unique fuel type appears.
- Returns the frequencies in descending order.

Example output:

```
HYBRID                     2792
GASOLINE                    611
FLEX FUEL                    89
COMPRESSED NATURAL GAS       27
```

---

## Result

The most common fuel type used by Chicago taxis is:

**HYBRID**
