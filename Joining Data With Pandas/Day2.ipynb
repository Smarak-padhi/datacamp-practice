# Exercise 1 — Counting Missing Rows with Left Join

## Problem

The `movies` and `financials` DataFrames are provided.

### Tasks

1. Select the best column to merge the two DataFrames.
2. Merge the `movies` DataFrame with the `financials` DataFrame using a **Left Join**.
3. Store the result in `movies_financials`.
4. Count the number of missing values in the `budget` column.
5. Print the result.

---

# Solution

```python
# Merge movies and financials using a Left Join
movies_financials = movies.merge(
    financials,
    on="id",
    how="left"
)

# Count missing values in the budget column
number_of_missing_fin = movies_financials["budget"].isna().sum()

# Print the result
print(number_of_missing_fin)
```

---

## Explanation

- Merge the tables using the `id` column.
- `how="left"` keeps all movies, even if they don't have matching financial data.
- `.isna()` finds missing (`NaN`) values.
- `.sum()` counts the missing values.

---

## Key Concepts

- `on="id"` → Merge using the common `id` column.
- `how="left"` → Keep all rows from the left DataFrame.
- `.isna()` → Check for missing values.
- `.sum()` → Count the missing values.

### Quick Tip

```python
df["column"].isna().sum()
```

Counts the number of missing values in a column.

# Chapter 2: Other Joins

---

# Right Join

A **Right Join** keeps **all rows from the right DataFrame**.

- Matching rows are added from the left DataFrame.
- If no match exists, missing values become `NaN`.

### Syntax

```python
df1.merge(
    df2,
    on="column_name",
    how="right"
)
```

---

# Left Join vs Right Join

| Left Join | Right Join |
|-----------|------------|
| Keeps all rows from the left table | Keeps all rows from the right table |
| Missing right values become `NaN` | Missing left values become `NaN` |

---

# Outer Join

An **Outer Join** keeps **all rows from both DataFrames**.

- Matching rows are combined.
- Non-matching rows are also kept.
- Missing values are filled with `NaN`.

### Syntax

```python
df1.merge(
    df2,
    on="column_name",
    how="outer"
)
```

---

# Inner vs Left vs Right vs Outer

| Join Type | Rows Kept |
|-----------|-----------|
| Inner | Only matching rows |
| Left | All rows from the left table |
| Right | All rows from the right table |
| Outer | All rows from both tables |

---

# Why Use an Outer Join?

Use an Outer Join when you don't want to lose any data.

Example:
- Finding records that exist in one table but not the other.
- Combining datasets while keeping every record.

---

# Visual Representation

```
Inner Join
Only the overlap

Left Join
All Left + Matching Right

Right Join
All Right + Matching Left

Outer Join
Everything
```

---

# Quick Summary

### Inner Join
- Keeps matching rows only.

### Left Join
- Keeps all rows from the left table.

### Right Join
- Keeps all rows from the right table.

### Outer Join
- Keeps all rows from both tables.

---

# Syntax Summary

```python
# Inner Join
df1.merge(df2, on="id")

# Left Join
df1.merge(df2, on="id", how="left")

# Right Join
df1.merge(df2, on="id", how="right")

# Outer Join
df1.merge(df2, on="id", how="outer")
```

---

# Key Takeaways

- `how="left"` → Keep all left rows.
- `how="right"` → Keep all right rows.
- `how="outer"` → Keep all rows from both tables.
- Missing matches are filled with `NaN`.
- If `how` is not specified, pandas performs an **Inner Join** by default.

# Exercise 3 — Right Join to Find Unique Movies

## Problem

The `movies`, `scifi_movies`, and `action_movies` DataFrames are provided.

### Tasks

1. Merge `action_movies` and `scifi_movies` using a **Right Join** on `movie_id`.
2. Add the suffixes:
   - `_act` for `action_movies`
   - `_sci` for `scifi_movies`
3. Keep only the rows where `genre_act` is missing (`NaN`).
4. Merge the result with `movies` using:
   - `id` (left table)
   - `movie_id` (right table)
5. Print the first few rows and the shape.

---

# Solution

```python
# Right Join
action_scifi = action_movies.merge(
    scifi_movies,
    on="movie_id",
    how="right",
    suffixes=("_act", "_sci")
)

# Keep only Sci-Fi movies
scifi_only = action_scifi[action_scifi["genre_act"].isna()]

# Merge with movies
movies_and_scifi_only = movies.merge(
    scifi_only,
    left_on="id",
    right_on="movie_id"
)

# Display the result
print(movies_and_scifi_only.head())
print(movies_and_scifi_only.shape)
```

---

## Key Concepts

- `how="right"` → Keep all rows from the **right** DataFrame.
- `.isna()` → Find missing values.
- `left_on` and `right_on` → Used when the merge columns have different names.

# Exercise 4 — Popular Genres with Right Join

## Problem

The `movie_to_genres` and `pop_movies` DataFrames are provided.

### Tasks

1. Merge `movie_to_genres` and `pop_movies` using a **Right Join**.
2. Match:
   - `movie_id` (left table)
   - `id` (right table)
3. Group by `genre`.
4. Count the number of movies (`id`) in each genre.
5. Plot a bar chart.

---

# Solution

```python
# Right Join
genres_movies = movie_to_genres.merge(
    pop_movies,
    how="right",
    left_on="movie_id",
    right_on="id"
)

# Count movies by genre
genre_count = genres_movies.groupby("genre").agg({"id": "count"})

# Plot
genre_count.plot(kind="bar")
plt.show()
```

---

## Key Concepts

- `how="right"` → Keep all rows from the right DataFrame.
- `left_on` and `right_on` → Merge columns with different names.
- `groupby()` → Group the data.
- `.agg({"id":"count"})` → Count the number of movies.
- `.plot(kind="bar")` → Create a bar chart.

# Exercise 5 — Using Outer Join to Select Actors

## Problem

The `iron_1_actors` and `iron_2_actors` DataFrames are provided.

### Tasks

1. Merge the two tables using an **Outer Join** on the `id` column.
2. Add suffixes:
   - `_1` for Iron Man 1
   - `_2` for Iron Man 2
3. Create a filter that selects rows where:
   - `name_1` is `NaN`, **or**
   - `name_2` is `NaN`
4. Print the first few rows.

---

# Solution

```python
# Outer Join
iron_1_and_2 = iron_1_actors.merge(
    iron_2_actors,
    on="id",
    how="outer",
    suffixes=("_1", "_2")
)

# Find actors that appear in only one movie
m = (
    iron_1_and_2["name_1"].isna() |
    iron_1_and_2["name_2"].isna()
)

# Display the result
print(iron_1_and_2[m].head())
```

---

## Key Concepts

- `how="outer"` → Keep all rows from both DataFrames.
- `.isna()` → Check for missing values.
- `|` → Logical **OR** operator.
- Rows with `NaN` indicate actors who appeared in **only one** of the movies.

# Chapter 2: Merging a Table to Itself (Self Join)

---

# What is a Self Join?

A **Self Join** means merging a DataFrame **with itself**.

It is useful when rows in the same table are related to each other.

---

# Why Use a Self Join?

Use a self join when you want to find relationships within the same DataFrame.

Examples:
- Employee → Manager
- Student → Mentor
- Parent → Child
- Flight → Connecting Flight

---

# Example Data

| id | name | manager_id |
|----|------|------------|
|1|Alice|3|
|2|Bob|3|
|3|Charlie|NaN|

Here, `manager_id` refers to another person's `id` in the same table.

---

# Self Join Syntax

```python
employees.merge(
    employees,
    left_on="manager_id",
    right_on="id",
    suffixes=("_emp", "_mgr")
)
```

---

# Why Use `left_on` and `right_on`?

Since both columns are in the same DataFrame, you tell pandas which columns to match.

```python
left_on="manager_id"
right_on="id"
```

Meaning:

Match each employee's `manager_id` with another employee's `id`.

---

# Using Suffixes

Both tables have the same column names.

Use suffixes to tell them apart.

```python
suffixes=("_emp", "_mgr")
```

Example:

```
name_emp
name_mgr
```

---

# Self Join with Different Join Types

### Inner Join

```python
employees.merge(
    employees,
    left_on="manager_id",
    right_on="id"
)
```

Returns only matching rows.

---

### Left Join

```python
employees.merge(
    employees,
    left_on="manager_id",
    right_on="id",
    how="left"
)
```

Keeps all rows from the left DataFrame.

---

# Common Uses

- Employee and Manager
- Parent and Child
- Student and Mentor
- Flight Connections
- Organizational Charts

---

# Key Takeaways

- A **Self Join** merges a DataFrame with itself.
- Use `left_on` and `right_on` when matching different columns.
- Use `suffixes` to distinguish duplicate column names.
- You can perform **Inner**, **Left**, **Right**, or **Outer** self joins.
